```python
# Convert all numerical strings to integers
# Replace None values with the average of the numerical values

def convert(data):
  y=[int(i) for i in data if i is not None]
  z=sum(y)//len(y)
  x=[int(i) if i is not None  else z for i in data ]
  for i in range(len(x)):
    if x[i]==0:
      x[i]=z
  return x

data = ["10", 20, None, "30", None, 50]
print(convert(data))


```

```python
# Write a function to check if two strings are anagrams (contain the same letters in a different order).

def sort(a):
  x=list(a)
  n=len(x)
  for i in range(n):
    for j in range(0,n-i-1):
      if x[j]>x[j+1]:
        x[j],x[j+1]=x[j+1],x[j]
  return x

def is_anagram(a,b):
  if sort(a)==sort(b):
    return 'it is anagram'
  else:
    return 'it is not'
print(is_anagram("listen", "silent"))



```

```python
# method overriding
class A:
  def mess(self):
    print("I am A")

class B(A):
  def mess(self):
    super().mess() #comment to override class A method
    print("I am B")
a=A()
b=B()
# a.mess()
b.mess()

# or
print("#########################################################")
class A:
  def __init__(self,name):
    self.name=name
  def det(self):
    print(f"I am {self.name}")
class B(A):
  def __init__(self,name,age):
    super().__init__(name)
    self.age=age
  def det(self):
    print(f"I am {self.name} and I am {self.age} years old")
a=A("Batman")
b=B('Hulk',30)
# a.det()
b.det()
```

```python
class Employee:
  company="Technospace"
  def __init__(self,name,age):
    self.name=name
    self.age=age
  @classmethod
  def setname(self,company):
    self.company=company
  @staticmethod
  def adult(age):
    if age>18:
      return True
    return False

a=Employee('James',19)
print(a.company)
a.setname("Acies")
print(a.company)
print(a.adult(19))
```
